```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['Goals', 'Assists', 'Shots on Target', 'Big Chances', 'Fouls Suffered']
markers = list(range(21))
str_markers = [str(i) for i in range(21)]  # Adjusted to match the range from 0 to 20

def make_radar_chart(name, stats, attribute_labels=labels,
                      plot_markers=markers, plot_str_markers=str_markers):

    labels = np.array(attribute_labels)

    angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)
    stats = np.concatenate((stats,[stats[0]]))  # Close the loop
    angles = np.concatenate((angles,[angles[0]]))  # Close the loop

    fig = plt.figure()
    ax = fig.add_subplot(111, polar=True)
    ax.plot(angles, stats, 'o-', linewidth=2)
    ax.fill(angles, stats, alpha=0.25)
    ax.set_thetagrids(np.degrees(angles[:-1]), labels)  # Adjusted to match the number of angles
    plt.yticks(plot_markers, plot_str_markers)  # Adjusting y-axis ticks
    ax.set_title(name)
    ax.grid(True)

    plt.show()

# Example usage:
make_radar_chart("Football Player Stats", [5, 2, 12, 4, 13])  # Pass the player's statistics as the second argument

```
